```python
#load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data=pd.read_csv('/content/Cardiotocographic.csv')
data
```

```python
# 1. Data Cleaning and Preparation
data.shape
```

```python
# Check data types
data.dtypes
```

```python
# Summary of a Dataframe
data.info()
```

```python
# Make a copy of dataset
data1=data.copy()
data1
```

```python
# Check for duplicates
data1[data1.duplicated()].shape
```

```python
data1[data1.duplicated()]
```

```python
# Drop Duplicates
data1_cleaned1=data1.drop_duplicates()
data1_cleaned1
```

```python
data1_cleaned1.shape
```

```python
# Check for missing values
missing_values = data1.isnull().sum()
missing_values
```

```python
data1_cleaned1.dtypes
```

```python
# Filling missing values with the median of each column
data1_cleaned1.fillna(data.median(), inplace=True)
data1_cleaned1.isnull().sum()
```

```python
# Detecting and handling outliers using the Interquartile Range (IQR) method
Q1 = data1_cleaned1.quantile(0.25)
Q3 = data1_cleaned1.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# Filtering outliers
lower_bound = Q1 - 1.5 * IQR
lower_bound
upper_bound = Q3 + 1.5 * IQR
upper_bound
data1_cleaned1_no_outlier=data1_cleaned1[~((data1_cleaned1 < (Q1 - 1.5 * IQR)) |(data1_cleaned1 > (Q3 + 1.5 * IQR))).any(axis=1)]
data1_cleaned1_no_outlier
data1_cleaned1_no_outlier.shape
```

```python
# 2. Statistical Summary
stats_summary = data1_cleaned1_no_outlier.describe()
stats_summary
```

```python
# Display mean, median, and standard deviation for key insights
mean_values = data1_cleaned1_no_outlier.mean()
median_values = data1_cleaned1_no_outlier.median()
std_values = data1_cleaned1_no_outlier.std()
print("Mean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nStandard Deviation Values:")
print(std_values)
```

```python
# 3. Data Visualization
# Histograms for distributions of numerical variables
data1_cleaned1_no_outlier.hist(bins=15, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
# Boxplots to identify distribution and potential outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=data1_cleaned1_no_outlier, palette="Set2")
plt.title("Boxplot of Variables")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data1_cleaned1_no_outlier.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Create the default pairplot
selected_features = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP']
sns.pairplot(data1_cleaned1_no_outlier[selected_features], diag_kind='kde')
plt.suptitle('Pairplot of Variables')
plt.show()
```

```python
# Advanced Visualization: Violin plots for variability
plt.figure(figsize=(15, 8))
sns.violinplot(data=data1_cleaned1_no_outlier[['ASTV', 'MSTV', 'ALTV', 'MLTV']], palette="pastel")
plt.xlabel("Variables")
plt.ylabel("Values")
plt.title("Violin Plot of Variables")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Pattern Recognition and Insights

# Identifying correlations between variables
correlations = data1_cleaned1_no_outlier.corr()
strong_corr = correlations[(correlations >= 0.5) | (correlations <= -0.5)].fillna(0)
strong_corr
```

```python
# Scatter plot to visualize the relationship between 'LB' and 'AC' (for example)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data1_cleaned1_no_outlier['LB'], y=data1_cleaned1_no_outlier['AC'])
plt.title("Scatter Plot of LB vs. AC")
plt.xlabel("Baseline Fetal Heart Rate (LB)")
plt.ylabel("Accelerations (AC)")
plt.show()
```
