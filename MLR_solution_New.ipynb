```python
#Preparing Model
```

```python
target=df1['Price']
```

```python
features = df1.drop(['Price'], axis=1)
features
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, train_size=0.8, random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
y_pred=linear_model.predict(x_test)
y_pred
```

```python
linear_model.score(x_test,y_test)
```

```python
linear_model.score(x_train,y_train)
```

```python
linear_model.coef_
```

```python
''' Interview Questions
Q1.1.What is Normalization & Standardization and how is it helpful?
Ans :
1.Normalization

a.It scales data to a fixed range, usually 0 to 1.
b.It shrinks big numbers and stretches small numbers to fit into the range.

Formula (Min-Max Normalization):

NormalizedÂ Value=(ð‘¥âˆ’min)/(maxâˆ’min)

* When to use it:

a.When you know the data has boundaries (like age between 0-100).
b.Important when algorithms rely on distances (like KNN, K-Means).

2.Standardization

a.It centers data around zero and scales it to have a standard deviation of 1.

b.The data becomes like a "standard" normal distribution.

Formula (Z-score Standardization): StandardizedÂ Value=(ð‘¥âˆ’mean)/standardÂ deviation

* When to use it:

a.When your data doesnâ€™t have a known range.
b.Useful when algorithms assume normal distribution (like Linear Regression, Logistic Regression, PCA).

* Why are they helpful?

a.Makes training faster: Features on similar scales help algorithms converge quickly.
b.Improves accuracy: Prevents large numbers from dominating small ones.
c.Essential for distance-based models: Like KNN, SVM, clustering without scaling, distance calculations are messed up.

Q2.What techniques can be used to address multicollinearity in multiple linear regression?
Ans :
Multicollinearity occurs when two or more independent variables in multiple linear regression are highly correlated with each other. This can cause issues in estimating the coefficients accurately, leading to unstable or unreliable results.

Here are some common techniques to address multicollinearity:
1. Remove Highly Correlated Variables
a.Simple solution: If two variables are highly correlated, you can drop one of them.
b.How to check: Use a correlation matrix or Variance Inflation Factor (VIF) to identify high correlations (VIF > 10 is often considered a red flag).

2. Combine Correlated Variables
Principal Component Analysis (PCA):
a.PCA transforms correlated variables into uncorrelated components.
b.This is a way to reduce dimensionality and deal with multicollinearity by combining variables into a smaller set of components.

3. Use Regularization Techniques
a.Ridge Regression (L2 Regularization): Adds a penalty to the size of the coefficients. This can reduce the impact of highly correlated variables and stabilize the regression.

b.Lasso Regression (L1 Regularization): Similar to Ridge, but can also shrink coefficients to zero. This is useful if you want to actually eliminate certain variables from the model.

4. Use Domain Knowledge
Use subject matter knowledge to decide which variables are important and potentially remove variables that are redundant or unnecessary.

5. Centering the Variables
Mean-centering: Subtract the mean of each variable.
This doesnâ€™t eliminate multicollinearity but can sometimes help to reduce its impact, especially in polynomial regression models.

6. Increase Sample Size
A larger dataset can help reduce the effect of multicollinearity. It provides more data to estimate the relationship between variables more accurately.

7. Check for Interaction Terms
Sometimes, interaction terms between variables (like multiplying two variables) can help explain relationships that might be lost in a standard linear model.
'''

```
